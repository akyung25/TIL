```
import requests
from bs4 import BeautifulSoup

url = 'https://comic.naver.com/webtoon/weekdayList.nhn?week=mon'

resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'lxml')
title_tag = soup.select('ul.img_list li dl dt a')
author_tag = soup.select('ul.img_list li dl dd a')
rating_tag = soup.select('ul.img_list li dl dd strong')

len(rating_tag)
```

```
data_list = []
for title, author, rating in zip(title_tag, author_tag, rating_tag):
  data = {
      'title' : title.text,
      'author' : author.text,
      'rating' : rating.text,
  }
  data_list.append(data)
```

```
data_list
```

```
for i in data_list:
  if i['author'] == '전체보기':
    del i

print(data_list)
```

과제2. rating이 9.9 이상인 웹툰의 제목 따로 recommended 변수에 담아주세요.

```
recommended = []

for i in data_list:
  if float(i['rating']) > 9.9:
    recommended.append(i['title'])

print(recommended)
```

```
def title_crawler(x):
  x = input('기사제목을 입력하세요:')
  url = 'https://search.naver.com/search.naver?where=news&query={}'.format(x)

  resp = requests.get(url)

  soup = BeautifulSoup(resp.content, 'lxml')
  a_tag = soup.select('ul.type01 li dl dt a')

  title = []
  for a in a_tag:
    title.append(a.text)

  return title
```

문제5. 인덱스를 기준으로 다음과 같이 인덱스에 맞는 문자열을 원소로 갖는 길이가 10인 리스트를 만들어 보세요.
idx = 0 -> zero
idx = 홀수 -> odd
idx = 짝수 -> even

['zero', 'odd', 'even', 'odd', 'even', 'odd', 'even', 'odd', 'even', 'odd']
